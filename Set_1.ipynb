```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Nielit_internship2025/SET_1.csv')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['month'], df['news_sold'], marker='o', linestyle='-', color='blue')
plt.title('News Sold per Month')
plt.xlabel('Month')
plt.ylabel('Number of News Sold')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

2. Add a new column to the DataFrame called "sales_category" that labels months as "High" if
news_sold is above 3000 and "Low" otherwise.

```python
df['sales_category'] = df['news_sold'].apply(lambda x: 'High' if x > 3000 else 'Low')
```

3. Sort the DataFrame in descending order based on the news_sold column and display the
result.

```python
sorted_df = df.sort_values(by='news_sold', ascending=False)
print("\nSorted DataFrame (by news_sold descending):")
print(sorted_df)
```

4. Calculate and print the average number of news sold across all months.

```python
average_news_sold = df['news_sold'].mean()
print(f"\nAverage News Sold: {average_news_sold:.2f}")
```

5. Find and display the month(s) with the maximum number of news sold.

```python
max_sold = df['news_sold'].max()
max_months = df[df['news_sold'] == max_sold]
print("\nMonth(s) with Maximum News Sold:")
print(max_months)
```
